**Q3. Explain the following terms**
* The difference in np.array, np.asarray and np.asanyarray
* The difference between Deep copy and shallow copy


* np.array : This function creates a numpy array from any valid Iterator object
* np.asarray : This function only creates creates a view of the array of how it will look if converted to array.
* np.asanyarray : Convert the input to an ndarray, but pass ndarray subclasses through.

* Deep Copy : On performing a deep copy operation over a data_item (eg: list,tuple,array etc) , Deep Copy actually creates a copy of the data item into the memory object instead of pointing out to the same data_item object into the memory.

* Shallow Copy : It only creates a pseudo copy of the data_item we wish to copy by just copying the reference of the data_item.

```python
#function to remove the non-numeric character from the entries of Phone Number
def non_num_rem(n):
  if(n!=np.nan):
    lst=list(filter((lambda x : x.isdigit()),str(n)))
    return "".join(lst)
```

```python
#Cleaning of Phone column
df["Phone"]=df["Phone"].apply(non_num_rem)

#Dataset after cleaning of phone column by removing all the non-numeric Character from Phone-Numbers
df.head(10)
```

```python
df["Phone"]=df["Phone"].apply(lambda x: '000' if(x=='')else x)  #Adjusting all the missing values and filling with 000 phone number
#Converting the Phone Column to numeric DataType
try:
  df["Phone"]=df["Phone"].astype(np.int64)
except ValueError:
  pass
  # df=df.astype({'Phone':np.nan})
df.head()
```

**Q15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:**

* a) Plot a histogram of the data with 30 bins.

* b) Overlay a line plot representing the normal distribution's probability density function (PDF).

* c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

* d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
#Numpy array of 1000 samples from a normal distribution

arr=np.random.randn(1000)

plt.title("Histogram with PDF Overlay")
plt.xlabel("Value");plt.ylabel("Frequency/Probability")
#a) Ploting a histogram
plt.hist(arr,bins=30,density=True)

#Generating x-axis value for the PDF line
x=np.linspace(min(arr),max(arr),1000)

#Generating the PDF function with respect to our Array:
pdf=((1/(np.std(arr)*np.sqrt(2*np.pi))))*np.exp((((x-np.mean(arr))/np.std(arr))**2)/-2)

#Plotiing the PDF line
plt.plot(x,pdf,color='m',linewidth=2.5)
plt.show()
```

**Q17.  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
#Creating two arrays each with 100 random data points
arr1,arr2=-10+(10-(-10))*np.random.rand(100),-10+(10-(-10))*np.random.rand(100)

plt.title("Quadrant-Wise SCatter Plot")
plt.xlabel("X-Axis");plt.ylabel("Y-Axis")

#Iterable object to assign a Categorical for Each Quadrants
quad=map((lambda x: "I" if(x[0]>=0 and x[1]>=0) else ("II" if(x[0]<0 and x[1]>=0) else ("III" if(x[0]<0 and x[1]<0) else "IV"))),zip(arr1,arr2))

sns.scatterplot(x=arr1,y=arr2,palette="gnuplot",hue=quad)  #Creation of the Scatter plot
plt.axvline(color='black',linewidth=1.7,linestyle="dashed")
plt.axhline(color='black',linewidth=1.7,linestyle="dashed")  #Creating the Quadrants

plt.legend(title="Quadrants")
plt.show()
```

**Q19.  Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'**

```python
x=np.random.randint(low=2,high=16,size=(50,))
category=tuple(map((lambda x: "Bad" if(x<=5) else ("Average" if(x>6 and x<=10) else "Good")),x))

p = figure(x_range=list(pd.Series(category).value_counts().index),height=350, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

p.vbar(x=pd.Series(category).value_counts().index, top=pd.Series(category).value_counts().values, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0
show(p)
```
